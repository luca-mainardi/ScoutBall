```python
import pandas as pd
```

```python
player_keepers = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_keepers.csv")
player_keepersadv = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_keepersadv.csv")
player_defense = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_defense.csv")
player_stats = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_stats.csv")
player_misc = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_misc.csv")
player_gca = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_gca.csv")
player_passing = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_passing.csv")
player_shooting = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_shooting.csv")
player_possession = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_possession.csv")
player_defense.head()
```

```python
defenders_data = player_defense[player_defense['position'] == 'DF'][['player', 'age', 'minutes_90s', 'team', 'position', 'tackles',
 'tackles_won', 'dribble_tackles_pct', 'blocked_passes', 'blocked_shots']]
defenders_data = pd.merge(defenders_data, player_stats[['player', 'club']], on='player', how='left')
defenders_data = pd.merge(defenders_data, player_misc[['player', 'aerials_won_pct']], on='player', how='left')
defenders_data = pd.merge(defenders_data, player_passing[['player', 'passes_pct']], on='player', how='left')
defenders_data['tackles'] = defenders_data['tackles_won'] / defenders_data['tackles'].replace(0, 1)
defenders_data['blocked_passes'] = defenders_data['blocked_passes'] / defenders_data['minutes_90s'].replace(0, 1)
defenders_data['blocked_shots'] = defenders_data['blocked_shots'] / defenders_data['minutes_90s'].replace(0, 1)
defenders_data = defenders_data.drop(columns=['tackles', 'tackles_won'])    
```

```python
defenders_data.head()
```

```python
midfielders_data = player_defense[player_defense['position'] == 'FW'][['player', 'age', 'minutes_90s', 'team', 'position', 'tackles', 
'tackles_won', 'interceptions']]
midfielders_data = pd.merge(midfielders_data, player_stats[['player', 'club', 'goals_assists_per90']], on='player', how='left')
midfielders_data = pd.merge(midfielders_data, player_passing[['player', 'passes_pct']], on='player', how='left')
midfielders_data = pd.merge(midfielders_data, player_possession[['player', 'dribbles_completed_pct']], on='player', how='left')
midfielders_data = pd.merge(midfielders_data, player_gca[['player', 'sca_per90']], on='player', how='left')
midfielders_data['tackles'] = midfielders_data['tackles_won'] / midfielders_data['tackles'].replace(0, 1)
midfielders_data['interceptions'] = midfielders_data['interceptions'] / midfielders_data['minutes_90s'].replace(0, 1)

midfielders_data = midfielders_data.drop(columns=['tackles_won'])    
```

```python
attackers_data = player_defense[player_defense['position'] == 'MF'][['player', 'age', 'minutes_90s', 'team', 'position', 'tackles_interceptions']]
attackers_data = pd.merge(attackers_data, player_stats[['player', 'club', 'goals_assists_per90']], on='player', how='left')
attackers_data = pd.merge(attackers_data, player_passing[['player', 'passes_pct']], on='player', how='left')
attackers_data = pd.merge(attackers_data, player_gca[['player', 'sca_per90']], on='player', how='left')
attackers_data = pd.merge(attackers_data, player_possession[['player', 'dribbles_completed_pct']], on='player', how='left')
attackers_data = pd.merge(attackers_data, player_shooting[['player', 'shots_on_target_pct']], on='player', how='left')
attackers_data['tackles_interceptions'] = attackers_data['tackles_interceptions'] / attackers_data['minutes_90s'].replace(0, 1)
```
